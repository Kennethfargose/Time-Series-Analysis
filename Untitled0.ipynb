```python
if 'date' in df.columns:
  df.loc[:, df.columns != 'date'] = df.loc[:, df.columns != 'date'].mask(df.loc[:, df.columns != 'date'] < 1000)
  df.interpolate(inplace=True)  # Interpolate the remaining NaN values
  # print(df.info)
  # Convert DataFrame to TimeSeries and name it based on the DataFrame name
  # ts_name = df_name + "_ts"  # Construct the TimeSeries name
ts = TimeSeries.from_dataframe(df, time_col='date', value_cols=[col for col in df.columns if col != "date"])
```

```python
def deep_learning_models( model, model_name,validation_start_date, model_kwargs,ts_list):
  transformer=Scaler()
  # Create an empty list to store the results
  res = []
  model = model(**model_kwargs)
  model_name = model_name

# Process each univariate TimeSeries
  for ts in ts_list:
      name = ts.components.item()
      ts_train, ts_test = ts.split_after(validation_start_date)
      ts_train = transformer.fit_transform(ts_train)
      ts_test_scaled = transformer.transform(ts_test)

      ts_val_scaled=ts_test_scaled[:194]
      ts_test_scaled=ts_test_scaled[:194]
      New_model = model.fit(series=ts_train,val_series=ts_val_scaled)
      preds = transformer.inverse_transform(New_model.predict(len(ts_test_scaled)))
      res.append(pd.DataFrame(
          {
          "model_name": [model_name],
          "ts_name": [name],
          #
          "mape": [mape(ts_test, preds)],
          "smape": [smape(ts_test, preds)],
          "r2": [r2_score(ts_test, preds)],
          }
      ))



  # Create a DataFrame from the results
  return pd.concat(res)
```

```python
MODEL_KWARGS={}
VALIDATION_START = pd.Timestamp("2014-10-18")

# Process the CSV files and get the results
nb= deep_learning_models(

            model=NBEATSModel,

            ts_list=[ts['CA_total']],

            model_name = "Nbeats",
            validation_start_date=VALIDATION_START,

            model_kwargs={"input_chunk_length": 30,
                  "output_chunk_length": 5,
                "generic_architecture":True,
                'num_stacks':30,
                "n_epochs":100,
              'activation':'ReLU',
                'dropout': 0,
                'layer_widths':256,
                'expansion_coefficient_dim':5,
                'trend_polynomial_degree':2,
                "force_reset": True,
                'num_blocks': 1,
                'num_layers': 1,
                'batch_size':37,

                'model_name':'Nbeats_final'

                  }


        )

```
